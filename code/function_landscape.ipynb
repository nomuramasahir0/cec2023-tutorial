```python
import numpy as np
import matplotlib as mpl
mpl.use("agg")
import matplotlib.cm as cm
import matplotlib.pyplot as plt
%matplotlib inline

mpl.rc("ps", useafm=True)  # force to use
mpl.rc("pdf", use14corefonts=True)  # only Type 1 fonts
mpl.rc("text", usetex=True)  # for a paper submission
```

```python
# sphere function
fig = plt.figure()
ax = plt.axes(projection="3d")
def sphere(x, y):
    return x ** 2 + y ** 2

x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)

X, Y = np.meshgrid(x, y)
Z = sphere(X, Y)

ax.plot_surface(X, Y, Z, cmap='winter')
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
# ax.set_title('Sphere')

fig.savefig("3d_sphere.png", bbox_inches='tight')

fig,ax=plt.subplots(1,1)
fig.set_figheight(3)
fig.set_figwidth(3)
cp = ax.contourf(X, Y, Z)
# ax.set_title("Sphere", fontsize=24)
plt.xticks([-3, 0, 3])
plt.yticks([-3, 0, 3])
plt.show()
fig.savefig("2d_sphere.png")
```

```python
# ellipsoid function
fig = plt.figure()
ax = plt.axes(projection="3d")
def ellipsoid(x, y):
    return x ** 2 + (1000*y) ** 2

x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)

X, Y = np.meshgrid(x, y)
Z = ellipsoid(X, Y)

ax.plot_surface(X, Y, Z, cmap='winter')
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
# ax.set_title('Ellipsoid')
fig.savefig("3d_ellipsoid.png", bbox_inches='tight')

fig,ax=plt.subplots(1,1)
fig.set_figheight(3)
fig.set_figwidth(3)
cp = ax.contourf(X, Y, Z)
# ax.set_title("Ellipsoid", fontsize=24)
plt.xticks([-3, 0, 3])
plt.yticks([-3, 0, 3])
plt.show()
fig.savefig("2d_ellipsoid.png")
```

```python
# rosen function
fig = plt.figure()
ax = plt.axes(projection="3d")
def rosen(x, y):
    return 100 * (y - x**2) ** 2 + (x-1) ** 2

x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)

X, Y = np.meshgrid(x, y)
Z = rosen(X, Y)

ax.plot_surface(X, Y, Z, cmap='winter')
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
# ax.set_title('Rosenbrock')
fig.savefig("3d_rosenbrock.png", bbox_inches='tight')

fig,ax=plt.subplots(1,1)
fig.set_figheight(3)
fig.set_figwidth(3)
cp = ax.contourf(X, Y, Z, 50)
# ax.set_title("Rosenbrock", fontsize=24)
plt.xticks([-2, 0, 2])
plt.yticks([-2, 0, 2])
plt.show()
fig.savefig("2d_rosenbrock.png")
```

```python
# rastrigin function
fig = plt.figure()
ax = plt.axes(projection="3d")
def rastrigin(x, y):
    return 10*2 + x**2 - 10*np.cos(2*np.pi*x) + y**2 - 10*np.cos(2*np.pi*y)

x = np.linspace(-3, 3, 30)
y = np.linspace(-3, 3, 30)

X, Y = np.meshgrid(x, y)
Z = rastrigin(X, Y)

ax.plot_surface(X, Y, Z, cmap='winter')
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
# ax.set_title('Rastrigin')
fig.savefig("3d_rastrigin.png", bbox_inches='tight')

fig,ax=plt.subplots(1,1)
fig.set_figheight(3)
fig.set_figwidth(3)
cp = ax.contourf(X, Y, Z)
# ax.set_title("Rastrigin", fontsize=24)
plt.xticks([-3, 0, 3])
plt.yticks([-3, 0, 3])
plt.show()
fig.savefig("2d_rastrigin.png")
```
